```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ====================== 1. 数据读取（从Excel文件读取） ======================
file_path = r"D:\project\稳定性测试.xlsx"
sheet_name = "AS7343-D65"

# 读取Excel数据
df = pd.read_excel(file_path, sheet_name=sheet_name)
# ====================== 2. 数据预处理 ======================
# 提取波长（通道名称）和测量数据
wavelengths = df.iloc[:, 0].values  # 所有波长（通道）
measure_data = df.iloc[:, 1:21].values  # 测量数据（行=通道，列=测量次数）

# 选取最后14组测量数据：第7~20次（列索引6~19）
start_col = 0  # 第7次测量的列索引
end_col = 14   # 第20次测量的列索引
measure_data_subset = measure_data[:, start_col:end_col]

# 最后14组的测量次数（横坐标：7-20次）
measure_times_subset = np.arange(start_col + 1, end_col + 1)

# ====================== 3. 绘图配置 ======================
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 创建画布（更大尺寸适配多通道曲线）
fig, ax = plt.subplots(figsize=(16, 9))

# 定义区分度高的颜色（按通道数量分配）
num_channels = len(wavelengths)
colors = plt.cm.tab10(np.linspace(0, 1, num_channels))  # tab10色系更适合通道区分

# ====================== 4. 绘制每个通道的测量曲线（仅最后14组） ======================
for i in range(num_channels):
    channel_name = f"{wavelengths[i]}"  # 通道名称（波长）
    # 该通道最后14组测量数据
    channel_data = measure_data_subset[i, :]
    
    # 绘制该通道的测量曲线
    ax.plot(
        measure_times_subset,    # 横坐标：7-20次测量
        channel_data,            # 纵坐标：该通道最后14次测量的值
        color=colors[i],
        label=channel_name,
        linewidth=2.0,
        alpha=0.8
    )
    
    # 绘制该通道的测量数据点
    ax.scatter(
        measure_times_subset,
        channel_data,
        color=colors[i],
        s=30,
        alpha=1.0,
        edgecolors='black',
        linewidths=0.5
    )

# ====================== 5. 图表美化 ======================
# 坐标轴标签
ax.set_xlabel('Times', fontsize=14, fontweight='bold')
ax.set_ylabel('Response Value', fontsize=14, fontweight='bold')

# # 标题（标注仅最后14组）
# ax.set_title('AS7343-D65 各波长通道第7~20次测量值变化', fontsize=16, fontweight='bold', pad=20)

# 横坐标刻度（强制显示7-20次，间隔1）
ax.set_xticks(measure_times_subset)
ax.set_xlim(measure_times_subset[0] - 0.5, measure_times_subset[-1] + 0.5)  # 左右留少量空白

# 网格线
ax.grid(True, linestyle='--', alpha=0.3, color='gray')

# # 图例（按波长通道显示，分多列避免过长）
# ax.legend(
#     bbox_to_anchor=(1.02, 1),
#     loc='upper left',
#     fontsize=10,
#     ncol=2,
#     frameon=True,
#     shadow=True,
#     title='波长通道',
#     title_fontsize=12
# )

# 坐标轴刻度美化
ax.tick_params(axis='both', labelsize=12)

# 调整布局
plt.tight_layout()

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ====================== 从Excel文件读取原始数据 ======================
file_path = r"D:\project\稳定性测试.xlsx"
sheet_name = "EVERFINE-D65"

# 读取Excel数据（确保数据类型为原始数值）
df = pd.read_excel(file_path, sheet_name=sheet_name, dtype='float64')


# ====================== 数据预处理 ======================
# 提取原始数据
wavelength_original = df['波长（nm)/数据ID'].values  # 波长列
measure_data_original = df.iloc[:, 1:21].values.astype(np.float64)  # 20次测量数据

# 1. 选取后14组数据（第7~20次，索引6~19）
start_idx = 0
end_idx = 14
measure_data_subset = measure_data_original[:, start_idx:end_idx]

# 2. 最大-最小归一化（Min-Max Scaling）：缩放到[0,1]区间
# 公式：normalized = (x - min) / (max - min)
data_min = measure_data_subset.min()  # 后14组数据的全局最小值
data_max = measure_data_subset.max()  # 后14组数据的全局最大值
measure_data_normalized = (measure_data_subset - data_min) / (data_max - data_min)

# ====================== 绘图 ======================
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(16, 9))
colors = plt.cm.Set3(np.linspace(0, 1, 20))  # 颜色匹配测量次数

# 绘制归一化后的后14组数据
for i in range(measure_data_normalized.shape[1]):
    # 归一化后的测量值
    y_meas = measure_data_normalized[:, i]
    # 对应原始测量次数（第7~20次）
    meas_num = start_idx + i + 1
    
    # 绘制归一化后的折线
    ax.plot(wavelength_original, y_meas, color=colors[start_idx + i], 
            label=f'第{meas_num}次', linewidth=1.6, alpha=0.6)
    
    # 绘制归一化后的数据点
    ax.scatter(wavelength_original, y_meas, color=colors[start_idx + i], s=18,
               alpha=1, edgecolors='black', linewidths=0.4)

# ====================== 纵轴设置（归一化后范围固定为[0,1]） ======================
ax.set_ylim(bottom=0, top=1)  # 归一化后范围严格[0,1]
ax.autoscale(enable=False, axis='y')  # 禁用自动缩放
ax.ticklabel_format(useOffset=False, style='plain', axis='y')  # 原始刻度格式

# ====================== 图表美化 ======================
ax.set_xlabel('Wavelength (nm)', fontsize=14, fontweight='bold')
ax.set_ylabel('SPD（Normalization）', fontsize=14, fontweight='bold')  # 标注归一化
# ax.set_title('EVERFINE-D65（第7~20次测量，Min-Max归一化）', 
#              fontsize=16, fontweight='bold', pad=20)

# ax.grid(True, linestyle='--', alpha=0.3)
# ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=9, ncol=2,
#           frameon=True, shadow=True, title='测量次数（第7~20次）', title_fontsize=11)
# ax.tick_params(axis='both', labelsize=12)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ====================== 从Excel文件读取原始数据 ======================
file_path = r"D:\project\稳定性测试.xlsx"
sheet_name = "EVERFINE-TL84"

# 读取Excel数据（确保数据类型为原始数值）
df = pd.read_excel(file_path, sheet_name=sheet_name, dtype='float64')

# ====================== 数据预处理 ======================
# 提取原始数据
wavelength_original = df['波长（nm)/数据ID'].values  # 波长列
measure_data_original = df.iloc[:, 1:21].values.astype(np.float64)  # 20次测量数据

# 1. 选取后14组数据（第7~20次，索引6~19）
start_idx = 0
end_idx = 14
measure_data_subset = measure_data_original[:, start_idx:end_idx]

# 2. 最大-最小归一化（Min-Max Scaling）：缩放到[0,1]区间
# 公式：normalized = (x - min) / (max - min)
data_min = measure_data_subset.min()  # 后14组数据的全局最小值
data_max = measure_data_subset.max()  # 后14组数据的全局最大值
measure_data_normalized = (measure_data_subset - data_min) / (data_max - data_min)

# ====================== 绘制原图表 ======================
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

fig, ax = plt.subplots(figsize=(16, 9))
colors = plt.cm.Set3(np.linspace(0, 1, 20))  # 颜色匹配测量次数

# 绘制归一化后的后14组数据
for i in range(measure_data_normalized.shape[1]):
    # 归一化后的测量值
    y_meas = measure_data_normalized[:, i]
    # 对应原始测量次数（第7~20次）
    meas_num = start_idx + i + 1
    
    # 绘制归一化后的折线
    ax.plot(wavelength_original, y_meas, color=colors[start_idx + i], 
            label=f'第{meas_num}次', linewidth=1.6, alpha=0.6)
    
    # 绘制归一化后的数据点
    ax.scatter(wavelength_original, y_meas, color=colors[start_idx + i], s=18,
               alpha=1, edgecolors='black', linewidths=0.4)

# ====================== 纵轴设置（归一化后范围固定为[0,1]） ======================
ax.set_ylim(bottom=0, top=1)  # 归一化后范围严格[0,1]
ax.autoscale(enable=False, axis='y')  # 禁用自动缩放
ax.ticklabel_format(useOffset=False, style='plain', axis='y')  # 原始刻度格式

# ====================== 图表美化 ======================
ax.set_xlabel('Wavelength (nm)', fontsize=14, fontweight='bold')
ax.set_ylabel('SPD（Normalization）', fontsize=14, fontweight='bold')  # 标注归一化
# ax.set_title('EVERFINE-TL84（第7~20次测量，Min-Max归一化）', 
#              fontsize=16, fontweight='bold', pad=20)

# ax.grid(True, linestyle='--', alpha=0.3)
# ax.legend(bbox_to_anchor=(1.02, 1), loc='upper left', fontsize=9, ncol=2,
#           frameon=True, shadow=True, title='测量次数（第7~20次）', title_fontsize=11)
# ax.tick_params(axis='both', labelsize=12)

plt.tight_layout()
plt.show()

# ====================== 单独生成竖排图例图片（核心新增逻辑） ======================
# 1. 计算图例数量和适配画布尺寸（竖排：宽度缩小，高度按图例数量自适应）
legend_num = measure_data_normalized.shape[1]  # 14个测量次数图例
fig_legend, ax_legend = plt.subplots(figsize=(4, legend_num * 0.8))  # 每个图例分配0.8英寸高度
ax_legend.axis('off')  # 关闭所有坐标轴元素，仅显示图例

# 2. 手动创建与原图表一致的图例项（保持颜色、标签、线样式一致）
legend_handles = []
for i in range(legend_num):
    meas_num = start_idx + i + 1
    # 创建与原折线样式匹配的图例手柄
    handle = plt.Line2D(
        [], [],
        color=colors[start_idx + i],
        linewidth=1.6,
        alpha=0.6,
        label=f'{meas_num}'
    )
    legend_handles.append(handle)

# 3. 添加竖排图例（ncol=1 实现竖排，居中显示）
legend = ax_legend.legend(
    handles=legend_handles,
    bbox_to_anchor=(0.5, 0.5),    # 图例在画布正中央
    loc='center',                 # 图例自身居中对齐
    fontsize=10,
    ncol=1,                       # 1列=竖排
    frameon=True,
    shadow=True,
    title='Times',
    title_fontsize=12,
    framealpha=1.0                # 图例背景不透明
)

# 4. 调整布局并保存高清图例图片
plt.tight_layout()
# 保存竖排图例（路径可自定义，dpi=300保证高清）
plt.savefig(r"D:\project\EVERFINE-TL84_竖排图例.png", dpi=300, bbox_inches='tight', facecolor='white')
plt.show()
```
